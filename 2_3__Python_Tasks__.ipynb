### Task 1: Greeting the User

### Task 2: Calculating Parcel Weight
This task calculates the total weight of souvenirs and trinkets based on their quantities and weights.

````python
# Task 2: Calculating Parcel Weight
# This function calculates the total weight of a parcel based on the number of souvenirs and trinkets.

def calculate_parcel_weight():
    # Taking input for the number of souvenirs and trinkets
    souvenirs = int(input("How many souvenirs do you have? "))
    trinkets = int(input("How many trinkets do you have? "))
    
    # Calculating the total weight
    total_weight = souvenirs * 75 + trinkets * 112  # weights in grams
    
    # Displaying the result
    print(f"The total weight of the parcel is: {total_weight} grams")

# Run the function
calculate_parcel_weight()
````

### Task 3: Calculating Compound Interest
This task calculates the compound interest for an initial deposit with a fixed annual interest rate over three years.

````python
# Task 3: Calculating Compound Interest
# This function calculates the compound interest for a deposit over 3 years at a 4% annual interest rate.

def calculate_interest():
    # Taking input for the initial deposit
    P = float(input("Enter your initial deposit (in USD): "))
    
    # Calculating and displaying the amount for each year
    for n in range(1, 4):  # Loop through years 1 to 3
        A = P * ((1 + 0.04) ** n)  # Compound interest formula
        print(f"Amount in the account at the end of year {n}: ${A:.2f}")

# Run the function
calculate_interest()
````

### Task 5: Coin Change Distribution
This task calculates the distribution of coins needed to make up a given amount of change.

````python
# Task 5: Coin Change Distribution
# This function calculates the number of coins needed to give a specific amount of change.

def calculate_change(change):
    # List of coin denominations (in cents)
    coins = [500, 200, 100, 50, 10, 5, 1]
    coin_count = {}

    # Calculate the number of each coin needed
    for coin in coins:
        if change >= coin:
            count = change // coin
            coin_count[coin] = count
            change -= count * coin

    return coin_count

def coin_change():
    # Taking input for the change amount
    change_input = int(input("Enter the change amount in cents: "))
    
    # Getting the coin distribution
    change_output = calculate_change(change_input)

    # Displaying the result
    print("To give the change, you will need:")
    for coin, count in change_output.items():
        if coin >= 100:
            print(f"{count} coin(s) of {coin // 100} dollars")
        else:
            print(f"{count} coin(s) of {coin} cents")

# Run the function
coin_change()
````

### Task 6: Bread Discount Calculation
This task calculates the total cost of buying discounted bread based on the number of loaves purchased.

````python
# Task 6: Bread Discount Calculation
# This function calculates the total cost of yesterday's bread with a discount.

def calculate_bread_cost():
    # Prices
    regular_price = 49.00  # Regular price per loaf in cents
    discount_price = regular_price * 0.4  # 60% discount
    
    # Displaying the prices
    print(f"Regular price per loaf: ${regular_price:.2f}")
    print(f"Price per yesterday's loaf with 60% discount: ${discount_price:.2f}")
    
    # Taking input for the number of loaves purchased
    quantity = int(input("Enter the number of yesterday's loaves purchased: "))
    
    # Calculating and displaying the total cost
    total_cost = quantity * discount_price
    print(f"Total cost of the bread: ${total_cost:.2f}")

# Run the function
calculate_bread_cost()
````
